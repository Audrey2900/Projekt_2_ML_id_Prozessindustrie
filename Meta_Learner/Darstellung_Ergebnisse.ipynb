```python
import matplotlib.pyplot as plt
import numpy as np

# --- Ergebnisse vorbereiten ---
steps = ["Step 1", "Step 3", "Step 7", "Step 8"]
metrics = ["Accuracy", "Recall", "F1-Score", "Precision"]

results = {
    "Step 1": [0.6020331888174615, 0.6124714828897339, 0.7515864128406122, 0.9724704177734846],
    "Step 3": [0.980981952260819, 1.0, 0.9903996865203762, 0.980981952260819],
    "Step 7": [0.9806763285024155, 1.0, 0.9902439024390245, 0.9806763285024155],
    "Step 8": [0.9841567351557803, 1.0, 0.9920151142480305, 0.9841567351557803],
}

# --- Farben für die Metriken ---
colors = ["royalblue", "orange", "green", "red"]

# --- Balken-Positionen ---
x = np.arange(len(metrics))  # Positionen für die Metriken
width = 0.2  # Breite der Balken

# --- Plot erstellen ---
fig, ax = plt.subplots(figsize=(10,6))

for i, step in enumerate(steps):
    values = results[step]  # Werte für diesen Step
    ax.bar(x + i*width - 1.5*width, values, width, label=step)

# --- Achsen und Labels ---
ax.set_ylabel("Score")
ax.set_ylim(0, 1.5)
ax.set_title("Modell-Evaluation pro Step")
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend(title="Step")
ax.axhline(y=1, linestyle='--', linewidth=2, alpha=0.4)

# --- Balkenwerte anzeigen ---
for i, step in enumerate(steps):
    for j, value in enumerate(results[step]):
        ax.text(x[j] + i*width - 1.5*width, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
